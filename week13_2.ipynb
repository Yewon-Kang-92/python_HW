(1) 저자 정보만 추출하여 데이터프레임 생성

```python
import requests
import pandas as pd

def get_author(isbn):
    url = f"https://www.googleapis.com/books/v1/volumes?q=isbn:{isbn}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        try:
            return ", ".join(data['items'][0]['volumeInfo']['authors'])
        except (KeyError, IndexError):
            return None
    else:
        return None

# 예시 ISBN 리스트
isbn_list = ["9788994492032", "9788936434267", "9788936433598"]

# 저자 정보 추출
author_data = []
for isbn in isbn_list:
    author = get_author(isbn)
    author_data.append({'ISBN': isbn, 'Author': author})

# 데이터프레임 생성
df_authors = pd.DataFrame(author_data)
df_authors
```

(2) 쪽수와 저자 정보를 함께 추출하여 두 개의 컬럼으로 저장

```python
def get_page_and_author(isbn):
    url = f"https://www.googleapis.com/books/v1/volumes?q=isbn:{isbn}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        try:
            info = data['items'][0]['volumeInfo']
            page = info.get('pageCount', None)
            author = ", ".join(info.get('authors', [None]))
            return page, author
        except (KeyError, IndexError):
            return None, None
    else:
        return None, None

# 페이지 수와 저자 동시 추출
combined_data = []
for isbn in isbn_list:
    page, author = get_page_and_author(isbn)
    combined_data.append({'ISBN': isbn, 'PageCount': page, 'Author': author})

# 데이터프레임 생성
df_combined = pd.DataFrame(combined_data)
df_combined
```
